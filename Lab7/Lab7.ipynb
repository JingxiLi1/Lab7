```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:\\Users\\51102\\Desktop\\Lab7\\onlinefoods.csv')
gender_counts = data['Gender'].value_counts()
fig, ax = plt.subplots()
ax.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Gender Distribution')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(data['Age'], bins=range(min(data['Age']), max(data['Age']) + 1, 1), edgecolor='black')
ax.set_title('Age Distribution')
ax.set_xlabel('Age')
ax.set_ylabel('Count')

plt.show()
```

```python
marital_status_counts = data['Marital Status'].value_counts()
fig, ax = plt.subplots()
ax.bar(marital_status_counts.index, marital_status_counts.values, color='skyblue')
ax.set_title('Marital Status Distribution')
ax.set_xlabel('Marital Status')
ax.set_ylabel('Count')

plt.xticks(rotation=45)  # Rotate labels to improve readability
plt.show()
```

```python
fig, ax = plt.subplots()
scatter = ax.scatter(data['longitude'], data['latitude'], alpha=0.5, c='red')
ax.set_title('Geographical Distribution of Respondents')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

plt.show()
```
